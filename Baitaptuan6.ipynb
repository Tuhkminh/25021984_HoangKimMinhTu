```python
#W6A1:Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
a=map(int,input().split())
ds=[]
for i in a:
    if i not in ds:
        ds.append(i)
print(ds)
```

```python
#W6A2:Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự
a=map(int,input().split())
ds=[]
n=0
for i in a:
    n+=i
    ds.append(n)
print(ds)
```

```python
#W6A3:Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuplesau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)
a=tuple(map(int,input().split()))
k=int(input("k="))
x=a[k:]+a[:k]
print(x)
```

```python
#W6A4:Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu keylặp lại, thêm vào list của key đó
d=input().split()
d1={}
for x in d:
    key,value=x.split(":")
    if key not in d1:
        d1[key]=[]
    d1[key].append(value)
print(d1)
```

```python
#W6A5:Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,'zeros': ...} (đếm số >0, <0, =0)
l=list(map(int,input().split()))
result={
    "positives":0,
    "negatives":0,
    "zeros":0}
for i in l:
    if i>0:
        result["positives"]+=1
    elif i<0:
        result["negatives"]+=1
    else:
        result["zeros"]+=1
print(result)
```

```python
#W6A8:Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiệncủa từng chuỗi và lưu vào dictionary.
l=input().split()
d={}
for s in l:
    if s in d:
        d[s]+=1
    else:
        d[s]=1
print(d)
```

```python
#W6A9:Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key. 
def ndict(x):
    s=input(x).split()
    d={}
    for item in s:
        key,value=item.split(":")
        d[key]=int(value)
    return d
d1=ndict("")
d2=ndict("")
d3={}
for key,value in d1.items():
    d3[key] = value
for key,value in d2.items():
    if key in d3:
        d3[key]+=value
    else:
        d3[key]=value
d4=dict(sorted(d3.items()))
print(d4)
```

```python
#W6A10:
l = list(map(int, input().split()))
k = int(input())
t=[]
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]+l[j]==k:
            t.append((l[i],l[j]))
t.sort(key=lambda x: x[0])
print(t)
```

```python
#W6A12:[Tìm phần tử xuất hiện nhiều nhất] 
l=list(map(int, input().split()))
d={}
for x in l:
    if x in d:
        d[x]+=1
    else:
        d[x]=1
maxi=0
for c in d.values():
    if c>maxi:
        maxi=c
l1=[]
for n in d:
    if d[n]==maxi:
        l1.append(n)
kq=min(l1)
print(kq)
```

```python
#W6A13:[Đảo ngược key và value của dictionary]
s=input().split()
d={}
for i in s:
    key,value=i.split(":")
    d[key]=value
d1={}
for key,value in d.items():
    d1[value]=key
print(d1)
```

```python
#W6A14:[Giao của hai danh sách]
l1=list(map(int, input().split()))
l2=list(map(int, input().split()))
c=[]
for x in l1:
    if x in l2 and x not in c:
        c.append(x)
print(c)
```

```python
#W6A15:[Lọc dictionary theo giá trị]
nd=input("").split()
k=int(input())
d={}
for i in nd:
    key,value=i.split(":")
    d[key]=int(value)
d1={}
for key,value in d.items():
    if value>k:
        d1[key]=value
print(d1)
```
