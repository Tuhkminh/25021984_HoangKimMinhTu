```python
#Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
h=int(input("Nhập chiều cao tam giác:"))
if h>1:
    for i in range(1,h+1):
        print(" "*(h-i),end="")
        if i==1:
            print("*")
        elif i==h:
            print("*"*(2*i-1))
        else:
            print("*"+" "*(2*i-3)+"*")
else:
    print("Nhập lại")
```

```python
#Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 
a=int(input("Nhập số nguyên dương a:"))
scp=False
for i in range(1,a+1):
    if i**2==a:
        scp=True
        break
if scp:
    print(f"{a} là số chính phương")
else:
    print(f"{a} không là số chính phương") 
```

```python
#Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 
a=int(input("Nhập số nguyên dương a:"))
if a>0:
    s=0
    for i in range(1,a):
        if a%i==0:
            s+=i
    if s==a:
        print(f"{a} là số hoàn hảo")
    else:
        print(f"{a} không là số hoàn hảo")
else:
    print("Nhập lại")
```

```python
#Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 
n=int(input("Nhập số nguyên dương n:"))
if n>1:
    k=0
    s=0
    while s+k+1<n:
        k+=1
        s+=k
    print(f"k={k}")
else:
    print("Nhập lại")
```

```python
#Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 
A=int(input("Nhập số nguyên dương A:"))
if A>1:
    n=1
    s=1
    while s<=A:
        n+=1
        s+=1/n
    print(f"n={n}")
else:
    print("Nhập lại")
```

```python
#Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng
maxi=None
mini=None
while True:
    n=int(input())
    if n==-1:
        break
    if maxi is None and mini is None:
        maxi=n
        mini=n
    else:
        if n>maxi:
            maxi=n
        if n<mini:
            mini=n
print("Số lớn nhất:",maxi)
print("Số nhỏ nhất:",mini)
```

```python
#Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều cách nhất có thể, ko dùng mảng.
n=int(input("Nhập số nguyên dương n:"))
if n>0:
    a=10
    x=1
    while n/a>=1:
        a*=10
        x+=1
    print("Số chữ số là",x)
else:
    print("Nhập lại ")
```

```python
#Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ
n=int(input("Nhập số nguyên dương n:"))
if n>0:
    c=0
    l=0
    while n>0:
        if n%2==0:
            c+=1
        else:
            l+=1
        n//=10
    print(f"{c} chữ số chẵn, {l} chữ số lẻ")
else:
    print("Nhập lại")
```

```python
#Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n
n=int(input("Nhập số nguyên dương n:"))
if n>0:
    a=n
    s=0
    while a>0:
        s+=(a%10)
        a//=10
    print(f"Tổng các chữ số của {n} là {s}")
else:
    print("Nhập lại")
```

```python
#Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 
n=int(input("Nhập số nguyên dương n:"))
if n>0:
    k=0
    while 3**k<n:
        k+=1
    if 3**k==n:
        print("là số dạng 3^k")
    else:
        print("Không là số dạng 3^k")
else:
    print("Nhập lại")
```

```python
# Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 
A=int(input("Nhập số nguyên dương A:"))
if n>0:
    x=1
    y=1
    while y<=A:
        x,y=y,y+x
    print(f"Số fibonacci lớn nhất không vượt quá {A} là {x}")
else:
    print("Nhập lại")
```

```python
#Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 
n=int(input("Nhập số nguyên dương n:"))
if n>0:
    a=n
    prime=0
    while a>0:
        x=a%10
        if x==2 or x==3 or x==5 or x==7:
            prime+=1
        a//=10
    print(f"{n} có {prime} chữ số nguyên tố")
else:
    print("Nhập lại")
```
