```python
#W6A16:Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣trận 2chiều, định dạng mỗi phần tử 4 vị trí căn phải
n,m = map(int,input().split())
mtr = []
for i in range(n):
    hang = list(map(int,input().split()))
    mtr.append(hang)
for i in range(n):
    for j in range(m):
        print(f"{mtr[i][j]:>4}",end="")
    print()
```

```python
# W6A18:Nhâp vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k) 
n,m,k = map(int,input().split())
mtr = []
for i in range(n):
    hang = list(map(int,input().split()))
    mtr.append(hang)
tong1 = 0
tong2 = 0
for i in range(n):
    tong1+=mtr[i][k-1]
for j in range(m):
    tong2+=mtr[k-1][j]
print(tong1)
print(tong2)
```

```python
#W6A19:Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list
lst = list(map(int, input().split()))
def vitrimax(lst):
    maxn = lst[0]
    vt = []
    for i in range(len(lst)):
        if lst[i] >= maxn:
            maxn = lst[i]
    for i in range(len(lst)):
        if lst[i] == maxn:
            vt.append(i+1)
    return vt
print(vitrimax(lst))
```

```python
#W6A20:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1
lst = list(map(int, input().split()))
k = int(input())
def vtk(lst):
    for i in range(len(lst)):
        if lst[i] == k:
            return i+1
print(vtk(lst))
```

```python
#W6A21:Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện
ch = input().split()
dic = {}
for i in range(len(ch)):
    x = ch[i]
    if x not in dic:
        dic[x] = []
    dic[x].append(i)
for x in dic:
    dic[x]=tuple(dic[x])
print(dic)
```

```python
#W6A23:VIết hàm xóa đi các số <x có trong mảng
lst = list(map(int, input().split()))
x = int(input())
def nho(lst):
    lst1 = []
    for i in range(len(lst)):
        if lst[i] >= x:
            lst1.append(lst[i])
    return lst1
print(nho(lst))
```

```python
#W6A24:Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau:
mylst = list(map(int,input().split()))
for i in range(len(mylst)):
    if mylst[i] == 0:
        del mylst[i]
        mylst.insert(0,0)
print(mylst)
```

```python
#W6A26:Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối
mylst = list(map(int,input().split()))
tong = 0
for i in range(1,len(mylst)-1):
    if mylst[i] > mylst[i-1] and mylst[i]>mylst[i+1]:
        tong += mylst[i]
print(tong)
```
