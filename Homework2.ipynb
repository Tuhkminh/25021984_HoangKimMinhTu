```python
#Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó.
a,b=map(float,input("Nhập 2 cạnh hình chữ nhật:").split())
if a>0 and b>0:
    print("Chu vi:",(a+b)*2)
    print("Diện tích:",a*b)
else:
    print("Cạnh không hợp lệ")
```

```python
#Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.
import math
r=float(input("Nhập bán kính hình tròn"))
if r>0:
    print("Diện tích:",math.pi*r**2)
    print("Chu vi:", math.pi*r*2)
else:
    print("Bán kính không hợp lệ")
```

```python
#Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.
import math
a,b,c=map(float,input("Nhập 3 số:").split())
if a>0 and b>0 and c>0 and a+b-c>1e-9 and a+c-b>1e-9 and b+c-a>1e-9:
    vuong=abs(a**2+b**2-c**2)<1e-9 or abs(a**2+c**2-b**2)<1e-9 or abs(b**2+c**2-a**2)<1e-9
    can=abs(a-b)<1e-9 or abs(a-c)<1e-9 or abs(b-c)<1e-9
    if abs(a-b)<1e-9 and abs(a-c)<1e-9 and abs(b-c)<1e-9:
        print("Tam giác đều")
    elif can:
        print("Tam giác cân")
    if vuong:
        print("Tam giác vuông")
    if not can and not vuong:
        print("Tam giác thường")
    print("Chu vi:",a+b+c)
    p=(a+b+c)/2
    print("Diện tích:",math.sqrt(p*(p-a)*(p-b)*(p-c)))
else:
    print("Không là 3 cạnh của một tam giác")
```

```python
#Bài 8: Giải hệ phương trình tuyến tính:
print("ax+by=m")
print("cx+dy=n")
a,b,m,c,d,n=map(float,input("a b m c d n :").split())
if a/c==b/d==m/n:
    print("Vô số nghiệm")
elif a/c==b/d and a/c!=m/n:
    print("Vô nghiệm")
else:
    print("x=",(m*d-b*n)/(a*d-b*c))
    print("y=",(a*n-m*c)/(a*d-b*c))
```

```python
#Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 
t=int(input("Thời gian:"))
if t>0:
    h=t//3600
    m=(t-h*3600)//60
    t=t-h*3600-m*60
    print(f"t={h} giờ {m} phút {t} giây")
else:
    print("Thời gian không hợp lệ")
```

```python
#Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
import math
print("Tâm I(x0,y0)")
x0=float(input("x0="))
y0=float(input("y0="))
r=float(input("Bán kính:"))
print("Điểm A(x,y)")
x=float(input("x="))
y=float(input("y="))
if r>0:
    if abs(math.sqrt((x-x0)**2+(y-y0)**2)-r)<1e-9:
        print("A(x,y) thuộc đường tròn")
    else:
        print("A(x,y) không thuộc đường tròn")
else:
    print("Bán kính không hợp lệ")
```
