```python
#[SumOfNumBers]
#Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để tính tống các số nguyên từ 1 đến n. (n <= 1000) 
try:
    n=int(input("Nhập vào một sô nguyên dương:"))
    if n>0:
        sum=0
        for i in range (1,n+1,1):
            print(i,end=" ")
            sum+=i
        print(f"\n{sum}")
    else:
        print("Nhập số âm rồi")
except ValueError:
    print("Bạn nhập không phải là một số nguyên")
```

```python
# [PrimeNumber] 
#VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. Kiểm tra số này có phải là số nguyên tố hay không? 
import math
n=int(input("Nhập một số nguyên dương:"))
while n<=0:
    n=int(input("Nhập lại một số nguyên dương:"))
if n<2:
    print("1 không phải số nguyên tố")
else:
    uoc=0
    for i in range(1,math.isqrt(n)+1):
        if n%i==0:
            uoc+=1
    if uoc>=2:
        print(f"{n} không phải số nguyên tố")
    else:
        print(f"{n} là số nguyên tố")
```

```python
#"[SumPrimesInRange]
#Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."
a,b=map(int,input("Nhập 2 số nguyên dương a,b:").split())
if 0<a<=b:
    def prime(n):
        if n<2:
            return False
        for i in range(2,n):
            if n%i==0:
                return False
        return True
    t=0
    for x in range(a,b+1):
        if prime(x):
            t+=x
    print(f"Tổng các số nguyên tố trong đoạn [{a},{b}]:",t)
else:
    print("Nhập lại")
```

```python
#"[LargestPrimeFactor]
#Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 
n=int(input("Nhập số nguyên dương n>=2:"))
if n>=2:
    i=2
    while n>1:
        if n%i==0:
            n//=i
        else:
            i+=1
    print("Ước số nguyên tố lớn nhất là:", i)
else:
    print("Nhập lại") 
```

```python
#ReverseAddPalindrome]
#Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được
n=int(input("Nhập số nguyên dương n:"))
buoc=0
while True:
    so=n
    reso=0
    while so>0:
        chuso=so%10
        reso=reso*10+chuso
        so//=10
    if n==reso:
        break
    n+=reso
    buoc+=1
print("Số bước:", buoc)
print("Giá trị palindrome:", n)
```

```python
#[PerfectSquaresDistinctDigits]
#Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).
n=int(input("Nhập vào số nguyên dương n:"))
a=n
import math
def scp(a):
    can=math.isqrt(a)
    return can*can==a
def khacnhau(a):
    while a>0:
        cuoi=a%10
        conlai=a//10
        while conlai>0:
            if conlai%10==cuoi:
                return False
            conlai//=10
        a//=10
    return True
co=False
for i in range(1,n+1):
    if scp(i) and khacnhau(i):
        print(i,end=" ")
        co=True
if not co:
    print("no number")
```

```python
#"[LongestCollatzUpToN]
#Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất." 
n=int(input("Nhập số nguyên dương n:"))
maxc= 0
maxx = 1
for i in range(1,n+1):
    x=i
    c=1
    while x!=1:
        if x%2==0:
            x=x//2
        else:
            x=3*x+1
        c=c+1
    if c>maxc:
        maxc=c
        maxx=i
print(f"Độ dài dãy Collatz lớn nhất là {maxc} với x={maxx}")
```

```python
#[EvenDivisors]
#Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6
n=int(input("Nhập số nguyên dương n<10^6:"))
count=0
for i in range(1,n+1):
    if n%i==0 and i%2==0:
        count+=1
print("Số lượng ước số chẵn của", n, "là:", count)
```

```python
#[ClosePairNumber]
#Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
def tonguoc(n):
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong+=i
    return tong
a = int(input("Nhập số thứ nguyên dương thứ nhất: "))
b = int(input("Nhập số thứ nguyên dương thứ hai: "))
if tonguoc(a) == b and tonguoc(b) == a:
    print("true")
else:
    print("false")
```

```python
#[GCD]
#Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n. 
m=int(input("Nhập số nguyên dương m:"))
n=int(input("Nhập số nguyên dương n:"))
if m>0 and n>0:
    u=1
    for i in range (1,min(m,n)+1):
        if m%i==0 and n%i==0:
            u=i
    print(f"Ước chung lớn nhất của {m} và {n} là {u}")
else:
    print("Nhập lại")
```

```python
# [ChickenAndDog]
#"Vừa gà vừa chó,Bó lại cho tròn,Ba mươi sáu con,Một trăm chân chẵn"
#Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid"
con=int(input("Nhập số con:"))
chan=int(input("Nhập số chân:"))
ga=0
cho=0
for i in range(1,con+1):
    if i*2+(con-i)*4==chan:
        ga=i
        cho=con-i
if ga+cho==con and ga>=0 and cho>=0:
    print(f"{ga} con gà và {cho} con chó")
else:
    print("invalid")
```
