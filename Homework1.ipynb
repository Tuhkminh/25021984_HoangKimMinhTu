```python
#2 Diện tích trồng hoa
a = float(input("chiều rộng a="))
b = float(input("chiều dài b="))
if 0 <a<b:
    S = a*b-3.14*(a/2)**2
    print("Diện tích trồng hoa:", int(S * 100) / 100)
else:
    print("nhập lại kích thước chính xác")
```

```python
#4 alphabet
c=input("Nhập 1 kí tự:")
if len(c)==1:
    if c.isalpha():
        print(f"{{{c}}} là kí tự alphabet")
    else:
        print(f"{{{c}}} không phải kí tự alphabet")
else:
    print("1 kí tự!")
```

```python
#5 chữ cái liền trước
c=input("Nhập 1 chữ cái in hoa:")
if len(c)==1 and c.isalpha() and c==c.upper():
    if c=="A":
        print("A là đặc biệt")
    else:
        code=ord(c.lower())-1
        chu=chr(code)
        print("Chữ cái in thường liền trước:",chu)
else:
    print("Nhập lại chính xác")
```

```python
#6 3 cạnh tam giác
import math
a,b,c=map(float,input("Nhập 3 số a,b,c:").split())
if a>0 and b>0 and c>0 and a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2
    S=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print("Là 3 cạnh tam giác với diện tích:",round(S,1))
else:
    print("Không phải 3 cạnh của tam giác")
```

```python
#8 Tiền điện
ten=input("Ten chu ho:")
trc=int(input("Chi so thang truoc:"))
nay=int(input("Chi so thang nay:"))
d=nay-trc
if 0<=d<=50:
    t=d*1984
elif d<=100:
    t=99200+(d-50)*2050
elif d<=200:
    t=201700+(d-100)*2380
elif d<=300:
    t=439700+(d-200)*2998
elif d<=400:
    t=739500+(d-300)*3350
else:
    t=1074500+(d-400)*3460
tien=t*1.08
print("Ho va ten:",ten)
print("Tien phai tra la:",int(tien))
```

```python
#thêm nhập ngày/tháng/năm hợp lệ
d,m,y=map(int,input("nhập ngày/ tháng/ năm:").split("/"))
nhuan=y%400==0 or (y%4==0 and y%100!=0)
match m:
    case 1|3|5|7|8|10|12:
        dmax=31
    case 4|6|9|11:
        dmax=30
    case 2:
        if nhuan:
            dmax=29
        else:
            dmax=28
if y>=1900 and 1<=m<=12 and 1<=d<=dmax:
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```
