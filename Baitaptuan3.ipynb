```python
#Kiểm tra xem một số có phải là lũy thừa của 2 hay không
#Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
#Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0
n=int(input("n="))
if n>0 and n&(n-1)==0:
    print("Là lũy thừa của 2")
else:
    print("Không là lũy thừa của 2")
```

```python
#Viêt chương trình chia một số m cho n, kết quả thu được làm tròn xuống
import math
m=float(input("m="))
n=float(input("n="))
if n!=0: 
    kq=math.floor(m/n)
    print("Kết quả làm tròn xuống là:",kq)
else:
    print("Không thể chia cho 0.")
```

```python
#[Even and Odd]
#Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ.
#Nếu x là số chẵn, in ra "Even". Nếu x là số lẻ, in ra "Odd". 
x=int(input("Số nguyên dương x="))
if x>0:
    if x%2==0:
        print("Even")
    else:
        print("Odd")
else:
    print("Nhập số nguyên dương")
```

```python
#[Triangle]
#Nhập vào 3 số nguyên dương a, b, c. Kiểm tra xem a, b, c có cấu thành độ dài cạnh của 1 tam giác được không. Nếu a, b, c cấu tạo thành được một tam giác, in ra "Yes". Ngược lại in ra "No"
a,b,c=map(int,input("Nhập 3 số nguyên dương a,b,c:").split())
if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0:
    print("Yes")
else:
    print("No")
```

```python
#[Largest number]
#Nhập vào 4 số nguyên. In ra màn hình số nguyên lớn nhất trong 4 số.
a,b,c,d=map(int,input("Nhập 4 số nguyên a,b,c,d:").split())
m=a
if b>m:m=b
if c>m:m=c
if d>m:m=d
print(f"Số lớn nhất:{m}")
```

```python
#[Phân loại tam giác]
#Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường".
a,b,c=map(int,input("Nhập 3 số nguyên dương a,b,c:").split())
if a+b>c and a+c>b and b+c>a and a>0 and b>0 and c>0:
    if a==b==c:
        print("Tam giác đều")
    elif a==b or a==c or b==c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

```python
#[Kiểm tra năm nhuận]
#Nhập vào một số nguyên dương n là số năm, kiểm tra xem năm đó có phải là năm nhuận hay không. Năm nhuận là năm chia hết cho 4 nhưng không chia hết cho 100, hoặc chia hết cho 400. Nếu là năm nhuận in ra "Yes", nếu không in ra "No"
y=int(input("Nhập năm:"))
if y>0:
    if (y%4==0 and y%100!=0) or y%400==0:
        print("Năm nhuận")
    else:
        print("Năm không nhuận")
else:
    print("Nhập lại")
```

```python
#[Giải phương trình bậc nhất]
#Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.
print("ax+b=0")
a=float(input("a="))
b=float(input("b="))
if a==0 and b==0:
    print("Vô số nghiệm")
if a==0 and b!=0:
    print("Vô nghiệm")
if a!=0:
    print("x=",round(-b/a,2))
```

```python
#[Xếp loại học lực]
#Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.
#Xếp loại học lực theo quy tắc:- >= 8.0: Giỏi- >= 6.5: Khá- >= 5.0: Trung bình- < 5.0: Yếu
d=float(input("Điểm trung bình:"))
if d>=8:
    print("Giỏi")
elif d>=6.5:
    print("Khá")
elif d>=5:
    print("Trung bình")
else:
    print("Yếu")
```

```python
#[Rounding]
#Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả cách nhau một dấu cách
a=float(input("Nhập số thực:"))
if a%1==0:
    a1=a
else:
    a1=a//1+1
a2=a//1
if abs(a1-a)>abs(a2-a):
    a3=a2
else:
    a3=a1
print(f"Kết quả làm tròn lên, tròn xuống, làm tròn tới số nguyên gần nhất:{a1} {a2} {a3}")
```
