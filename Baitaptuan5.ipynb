```python
#W5A1:Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two(a, b) -> int: 
a,b=map(int,input("Nhập 2 số nguyên a,b: ").split())
def Max_Of_Two(a,b):
    if a>=b:
        return a
    else:
        return b
print(Max_Of_Two(a,b))
```

```python
#W5A3: Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
import math
a=int(input("Nhập số nguyên a:"))
def snt(a):
    if a<2:
        return False
    else:
        for i in range(2,math.isqrt(a)+1):
            if a%i==0:
                return False
        return True
if snt(a):
    print("Số nguyên tố")
else:
    print("Không là số nguyên tố")
```

```python
#W5A4:VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên
a=int(input("Nhập số nguyên a:"))
def shh(a): 
    n=0
    for i in range (1,a): 
        if a%i==0: 
            n+=i 
    return n==a 
if shh(a):
    print("Số hoàn hảo")
else:
    print("Không phải số hoàn hảo")
```

```python
#W5A5:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí củaphần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 
def vt(ds,k):
    for i in range(len(ds)):
        if ds[i]==k:
            return i
    return -1
ds=list(map(int, input("Nhập danh sách số nguyên:").split()))
k=int(input("Nhập số nguyên dương k:"))
vt=vt(ds,k)
print(f"Vị trí phần tử đầu tiên bằng {k} là: {vt+1}")
```

```python
#W5A6: Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)
def gt(n):
    a=1
    if n==0:
        return 1
    else:
        for i in range(1,n+1):
            a*=i
        return a
n=int(input("Nhập số nguyên n:"))
print(f"n!={gt(n)}")
```

```python
#W5A7
def mtbt(num1,num2,operat):
    if operat=="+":
        kq=num1+num2
    elif operat=="-":
        kq=num1-num2
    elif operat=="*":
        kq=num1*num2
    elif operat=="/":
        if num2==0:
            return "Lỗi"
        else:
            kq=num1/num2
    else:
        return "Lỗi"
    return round(kq,2)
num1=float(input("Nhập số thứ nhất: "))
num2=float(input("Nhập số thứ hai: "))
operat=input("Nhập toán tử:")
print("Kết quả:",mtbt(num1,num2,operat))
```

```python
#W5A8:Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
def ham(x,y):
    n=0
    while x>0 or y>0:
        bitx=x%2
        bity=y%2
        if bitx!=bity:
            n+=1
        x//=2
        y//=2
    return n
x=int(input("Nhập số nguyên x: "))
y=int(input("Nhập số nguyên y: "))
print(f"Khoảng cách Hamming giữa {x} và {y} là: {ham(x,y)}")
```

```python
#W5A9:Viết hàm trả về tổng các chữ số của số nguyên dương n 
def tcs(n):
    i=0
    while n>0:
        i+=n%10
        n//=10
    return i
n=int(input("Nhập số nguyên dương n:"))
print(f"Tổng các chữ số của {n} là: {tcs(n)}")
```

```python
#W5A10
def dang_cau(a, b):
    for i in range(len(a)):
        if a.index(a[i]) != b.index(b[i]):
            return False
    return True
word1=input("Nhập từ thứ nhất: ")
word2=input("Nhập từ thứ hai: ")
if dang_cau(word1, word2):
    print("True")
else:
    print("False")
```
