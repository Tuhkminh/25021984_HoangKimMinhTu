```python
#W11A1
from pathlib import Path
WORKDIR = Path(".")
path=input("Moi ban nhap ten file:")
def reverse(path: str) -> int:
    file = WORKDIR/path
    with open(file,"r") as f:
        n=f.read().strip()
    reversed_num = int(n[::-1])
    return reversed_num
print(reverse(path))
```

```python
#W11A2
from pathlib import Path
WORKDIR = Path(".")
path=input("Moi ban nhap ten file:")
def canWinNim(path:str)-> bool:
    file = WORKDIR/path
    with open(file,"r") as f:
        n=int(f.read().strip())
    if n%4==0:
        return False
    else:
        return True
if (canWinNim(path)):
    print("Toi thang roi")
else:
    print("Ban thang roi")
print("Passed test")
```

```python
#W11A3
from pathlib import Path
WORKDIR = Path(".")
path=input("Moi ban nhap ten file:")
def housesOfHogwarts(path: str):
    file = WORKDIR/path
    with open(file,"r") as f:
        n=int(f.readline().strip())
        for i in range(n):
                name = f.readline().strip()
                if name:
                    print(name)
housesOfHogwarts(path)
```

```python
#W11A4:
from pathlib import Path
WORKDIR = Path(".")
path1=input("Moi ban nhap ten file 1:")
path2=input("Moi ban nhap ten file 2:")
def areAnagrams(path1: str, path2: str):
    file = WORKDIR/path1
    with open(file,"r") as f:
        n1=f.readline().strip()
    file = WORKDIR/path2
    with open(file,"r") as f:
        n2=f.readline().strip()
        n1=n1.lower().replace(" ","")
        n2=n2.lower().replace(" ","")
        d1={}
        d2={}
        for i in n1:
            if i not in d1:
                d1[i]=1
            else:
                d1[i]+=1
        for i in n2:
            if i not in d2:
                d2[i]=1
            else:
                d2[i]+=1
        for i in d1:
            if d1[i]!=d2.get(i,0):
                return "false"
        return "true"
print(areAnagrams(path1, path2))
```
