```python
#W7A1:Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần Trả về chỉ số của phần tử hoặc -1 nếu không tìmthấy. mẫu hàm: def binary_search(arr, left, right, target)
def binary_search(arr, left, right, target):
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, left, mid - 1, target)
    else:
        return binary_search(arr, mid + 1, right, target)
lst = list(map(int, input().split()))
target = int(input())
print(binary_search(lst, 0, len(lst) - 1, target))
```

```python
#W7A2:Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần trong danh sách. def count_occurrences(arr, x) -> int:
arr = list(map(int,input().split()))
x = int(input())
def count_occurrences(arr, x):
    n = 0
    for i in arr:
        if i == x:
            n += 1
    return n            
print(count_occurrences(arr, x))
```

```python
#W7A3:Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort);Sắp xếp chọn (Selection Sort)
lst = list(map(int,input().split()))
def bubble_sort():
    for i in range(1,len(lst)):
        if lst[i-1] >= lst[i]:
            lst[i-1],lst[i] = lst[i],lst[i-1]
            return bubble_sort()
    return lst
print(bubble_sort())
def selection_sort():
    n = len(lst)
    for i in range(n - 1):
        mini = i
        for j in range(i + 1, n):
            if lst[j] < lst[mini]:
                mini = j
        lst[i], lst[mini] = lst[mini], lst[i]
    return lst
print(selection_sort())
```

```python
#W7A4: Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.
lst = list(map(int,input().split()))
def maxpt():
    maxs = lst[0]
    maxn = 0
    for i in range(len(lst)):
        if lst.count(lst[i]) > maxn:
            maxn = lst.count(lst[i])
            maxs = lst[i]
    return(f"{maxs} xuat hien nhieu nhat, som nhat, {maxn} lan")
maxpt()
```

```python
#W7A5: Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao cho a + b = X
lst = list(map(int,input().split()))
X = int(input())
def demso():
    n = 0
    a = 1
    while a <= len(lst):
        for i in range(a,len(lst)):
            if lst[a-1] + lst[i] == X :
                n += 1
        a += 1
    return n
print(demso())
```

```python
#W7A7: Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề)
nums = list(map(int,input().split()))
n = len(nums)
dc = []
for i in range(len(nums)):
    dc.append(1)
    for j in range(i):
        if nums[j] < nums[i]:
            dc[i] = max(dc[i], dc[j] + 1)
print(max(dc))
```
