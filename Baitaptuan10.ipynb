```python
#W9A1
a=list(map(int,input().split()))
b=list(map(int,input().split()))
def merge(a,b):
    c=[]
    i=0
    j=0
    while i<len(a) and j<len(b):
        if a[i]<b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1 
    c=c+a[i:]+b[j:]
    return c
print(merge(a,b))
```

```python
#W9A4
def xau(a,b):
    possiblej={0}
    for chara in a:
        nextpossiblej=set()
        for j in possiblej:
            if chara.islower():
                nextpossiblej.add(j)
                if j<len(b) and chara.upper()==b[j]:
                    nextpossiblej.add(j+1)
            else:
                if j<len(b) and chara==b[j]:
                    nextpossiblej.add(j+1)
        possiblej=nextpossiblej
        if not possiblej:
            break
    if len(b) in possiblej:
        return "YES"
    return "NO"
a=input()
b=input()
print(xau(a,b))
```

```python
#W9A5
def kiemtraday(l1,l2):
    for i in range(len(l2)-len(l1)+1):
        l3= l2[i : i + len(l1)]    
        if l3==l1:
            return "YES"
    return "NO"
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
print(kiemtraday(l1,l2))
```

```python
#W9A7
def canPlaceFlowers(flowerbed,k):
    for i in range(len(flowerbed)):
        if flowerbed[i]==0:
            left=(i == 0) or (flowerbed[i - 1] == 0)
            right=(i==len(flowerbed)-1) or (flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i]=1
                k-=1
    return k<=0
flowerbed=list(map(int,input().split()))
k=int(input())
print(canPlaceFlowers(flowerbed,k))
```

```python
#W9A8
def is_path_crossing(moves):
    l=[[0,0]]
    x=0
    y=0
    for i in moves:
        if i=="U":
            y+=1  
        if i=="D":
            y-=1
        if i=="R":
            x+=1
        if i=="L":
            x-=1
        a=[x,y]
        if a not in l:
            l.append(a)
        else:
            return True
    return False
moves=input()
print(is_path_crossing(moves))
```

```python
#W9A10:
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    maxr=0
    i=0
    for house in houses:
        while i+1<len(heaters) and abs(heaters[i+1]-house)<=abs(heaters[i]-house):
            i+=1
        r = abs(heaters[i]-house)
        if r>maxr:
            maxr=r
    return maxr
houses=list(map(int,input().split()))
heaters=list(map(int,input().split()))
print(findRadius(houses, heaters))
```
