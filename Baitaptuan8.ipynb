```python
#W7A16:
lst = list(map(int,input().split()))
new =[]
for i in lst:
    if i == lst.count(i):
        new.append(i)
    if i < lst.count(i) and i > new.count(i):
        new.append(i)
if lst == new:
    print("YES")
else:
    print(f"NO {len(lst)-len(new)}") 
```

```python
#W7A18:
lst = list(map(int,input().split()))
chan = []
le = []
for i in lst:
    if i % 2 == 0:
        chan.append(i)
    else:
        le.append(i)
def sxchan(a):
    for i in range(len(chan)):
        for j in range(len(chan)-1):
            if chan[j] > chan[j+1]:
                chan[j], chan[j+1] = chan[j+1], chan[j]
    return a
def sxle(a):
    for i in range(len(le)):
        for j in range(len(le)-1):
            if le[j] < le[j+1]:
                le[j], le[j+1] = le[j+1], le[j]
    return a
chan1 = sxchan(chan)
le1 = sxle(le)
new = chan1 + le1
print(new)
```

```python
#W7A21:
N, T = map(int,input().split())
minN = 1
tong = 0
match T:
    case 1 | 3 | 5 | 7 | 8 | 10 | 12:
        maxn = 31
    case 4 | 6 | 9 | 11:
        maxn = 30
    case 2:
        maxn = 29
if 1 <= T <= 12 and 1 <= N <= maxn:
    for i in range(1, T):
        match i:
            case 1 | 3 | 5 | 7 | 8 | 10 | 12:
                tong += 31
            case 4 | 6 | 9 | 11:
                tong += 30
            case 2:
                tong += 29
    tong += N
    print(tong)
else:
    print("invalid")
```

```python
#W7A22:
N = int(input())
hovaten=[]
ten=[]
for i in range(N):
    hvt = input()
    hovaten.append(hvt)
for i in hovaten:
    chu = i.split()
    t = chu[-1]
    ten.append(t)
dem = {}
for i in ten:
    if i in dem:
        dem[i] += 1
    else:
        dem[i] = 1
t = ""
slmax = 0
for i in dem:
    sl = dem[i]
    if sl > slmax:
        slmax = sl
        t = i
    elif sl == slmax:
        if t == "" or i < t:
            t = i
print(t)
```

```python
#W7A31
so = int(input())
s = list(str(so))
n = len(s)
smin = int(so)
for i in range(n):
    for j in range(i + 1, n):
        s[i], s[j] = s[j], s[i]
        if s[0] != '0':
            new = int("".join(s))
            if new < smin:
                smin = new
        s[i], s[j] = s[j], s[i]
print(smin)
```
